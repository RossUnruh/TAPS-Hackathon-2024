```python
# soil volumetric water content time series. This is a team wise graph.
plt.figure(figsize=(18,5))
plt.plot (moisture_team_3["Timestamp"], moisture_team_3["VWC"])
plt.ylim(0.3, 0.5)
plt.xlabel("Date", fontsize = 16)
plt.ylabel("Profile Volumetric Water Content( m/m)", fontsize = 16)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.show()
```

```python
# Temperature This is a common one for whole field. 
mesonet = pd.read_excel(r"C:\\Priyanka\\hackthone-kansas\\Water-squad-dash\\dataset\\colby_station_kansas_mesonet_with_ETO.xlsx")
mesonet["TIMESTAMP"] = pd.to_datetime(mesonet["TIMESTAMP"])
plt.figure(figsize=(18,5))
plt.plot (mesonet["TIMESTAMP"], mesonet["TEMP2MAVG"])
plt.xlabel("Date", fontsize = 16)

plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.ylabel("Average Temperature (\u00B0C)", fontsize = 16)
plt.show()
```

```python
total_precip = (round(mesonet["PRECIP"].sum(), 3))
print(total_precip)
daily_precip = mesonet.groupby("TIMESTAMP")["PRECIP"].sum()
print(daily_precip)
```

```python
# Precipitation 
plt.figure(figsize=(18,5))
plt.bar (mesonet["TIMESTAMP"], mesonet["PRECIP"])
plt.xlabel("Date", fontsize = 16)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.ylabel("Precipitation (mm)", fontsize = 16)
plt.text(daily_precip.index[-50], 35, f"Total precipitation: {total_precip} mm", fontsize=15)
plt.show()
```

```python
# Average relative humidity
plt.figure(figsize=(18,5))
plt.plot (mesonet["TIMESTAMP"], mesonet["RELHUM2MAVG"] )
plt.xlabel("Date", fontsize = 16)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.ylabel("Average Relative Humidity (%)", fontsize = 16)
plt.show()
```
