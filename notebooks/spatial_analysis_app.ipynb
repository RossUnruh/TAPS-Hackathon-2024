```python
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image

#NDVI based 

# Path to the folder containing images
IMAGE_FOLDER = Path(r'C:\Priyanka\hackthone-kansas\Water-squad-dash\dataset\NDVI images')

# Get the first image in the folder
image_files = list(IMAGE_FOLDER.glob("*.jpg"))  # Change "*.png" if your images have a different extension
if image_files:
    first_image_path = image_files[0]
    
    # Open and display the image
    img = Image.open(first_image_path)
    plt.imshow(img)
    plt.axis('off')  # Hide axis
    plt.title(f"Displaying: {first_image_path.name}")
    plt.show()
else:
    print("No images found in the specified folder.")
```
